#### Important Note

For this exercise, you MUST use the following KerasHub components:

- [keras_hub.models.ViTImageClassifier](https://keras.io/keras_hub/api/models/vit/vit_image_classifier/)

- [keras_hub.models.ViTImageClassifierPreprocessor](https://keras.io/keras_hub/api/models/vit/vit_image_classifier_preprocessor/)

This ensures your input preprocessing (resizing + normalization) matches what the pretrained ViT model expects.

Do not replace the preprocessor with manual normalization (such as dividing by 255), because it may produce incorrect predictions.

```python
# Import Libraries
import matplotlib.pyplot as plt
from PIL import Image
import requests

import tensorflow as tf
import keras_hub 
import numpy as np

# Load ViTImageClassifierPreprocessor (vit_base_patch16_224_imagenet preset)

preprocessor = keras_hub.models.ViTImageClassifierPreprocessor.from_preset(
    "vit_base_patch16_224_imagenet"
)

# Load ViTImageClassifier (vit_base_patch16_224_imagenet preset)
model = keras_hub.models.ViTImageClassifier.from_preset(
    "vit_base_patch16_224_imagenet"
)

```

```python
# Load the images
image_paths = [
    "/content/lab03-image-classification-vision-transformer-vit/images/cat.jpg",
    "/content/lab03-image-classification-vision-transformer-vit/images/dog.jpg",
    "/content/lab03-image-classification-vision-transformer-vit/images/car.jpg",
    "/content/lab03-image-classification-vision-transformer-vit/images/banana.jpg",
    "/content/lab03-image-classification-vision-transformer-vit/images/airliner.jpg"
]

true_labels = [
    "tabby cat",
    "golden retriever",
    "sports car",
    "banana",
    "airliner"
]
from PIL import Image

images = []

for path in image_paths:
    img = Image.open(path).convert("RGB")
    img = img.resize((224, 224))   
    img_array = np.array(img)
    images.append(img_array)

image_batch = np.array(images)

image_batch = tf.convert_to_tensor(image_batch)

```

```python
processed_images = preprocessor(image_batch)
```

```python
class_url = "https://raw.githubusercontent.com/pytorch/hub/master/imagenet_classes.txt"
classes = requests.get(class_url).text.splitlines()

# Predict classes
predictions = model(processed_images)
predicted_ids = tf.argmax(predictions, axis=1).numpy()
predicted_labels = [classes[i] for i in predicted_ids]

for i, label in enumerate(predicted_labels):
    print(f"Image {i+1}: {label}")
```
